```python
import cv2
from matplotlib import pyplot as plt
```

```python
img_path = "./Lena-2019-CMS-Source2_WEB.jpg"
```

#### Load and Read images

```python
img = cv2.imread(img_path,cv2.IMREAD_UNCHANGED)
#img2 = cv2.imread(img_path2, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img)
plt.title('img')
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_img)
plt.title('img')
plt.show()
```

#### Split color channel into IMREAD_COLOR

```python
(R,G,B) = cv2.split(RGB_img)#提取R、G、B分量
```

```python
cv2.imshow("color_lena", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
